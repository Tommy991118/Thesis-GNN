## Notebook containing figures used in the final version of the thesis and additional figures

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

Load in data

```python
df = pd.read_csv("Results.csv")
df_school = pd.read_csv("Results_School.csv")
mape = pd.read_csv("MAPE.csv")
mape_school = pd.read_csv("MAPE_school.csv")
```

Clean data to be more representable

```python
df['value'] = df['value'].str.replace(',','.')
df['value'] = df['value'].astype(float)
df_school['value'] = df_school['value'].str.replace(',','.')
df_school['value'] = df_school['value'].astype(float)
mape['value'] = mape['value'].str.replace(',','.')
mape['value'] = mape['value'].astype(float)
mape_school['value'] = mape_school['value'].str.replace(',','.')
mape_school['value'] = mape_school['value'].astype(float)
```

```python
df_test = df.loc[df['test or train'] == 'Test']
df_train = df.loc[df['test or train'] == 'Train']
df_school_test = df_school.loc[df_school['test or train'] == 'Test']
df_school_train = df_school.loc[df_school['test or train'] == 'Train']
df_test = df_test.sort_values(by = 'metric', axis = 0)
```

Use filtering based on different metrics in order to build seperate figures

```python
df_test_RMSE = df_test.loc[df['metric']== 'RMSE']
df_test_MAE = df_test.loc[df['metric']== 'MAE']
df_test_MAPE = df_test.loc[df['metric']== 'MAPE']
```

### Start section with figures house-house embedding

RMSE results of test set

```python
graph_data = df_test_RMSE.loc[df['dataset'].isin(['H-H emb + Raw (with lat & long)','Raw Base (with lat & long)'])]
graph_data = graph_data.sort_values(by =  'dataset', axis = 0, ascending = False)

g =sns.catplot(
    data=graph_data, kind="bar",
    x="model", y="value", hue="dataset", palette='colorblind', alpha=0.8, width = 0.9, height=4, aspect=6/4, ci=None)
g.set_axis_labels("Model", "Value")
g.set(title = 'RMSE-values')
g.legend.set_title("Results on test set") 

for ax in g.axes.ravel():
    for c in ax.containers:
        labels = [f'{(v.get_height()):.4f}' for v in c]
        ax.bar_label(c, labels= labels, label_type='edge')
```
